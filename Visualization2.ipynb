```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
emp = {'Age':[10,23,22,25,20,23,24,27,25,24,33,21],
       'Salary':[15,22,24,25,30,23,34,43,19,20,52,21]}

df = pd.DataFrame(data=emp)

df
```

```python
plt.figure(figsize=(10,8))

ax = sns.boxplot(data=df)
plt.yticks(range(5,60,5))
plt.xlabel('Age and Salary')

plt.show()
```

```python
df.plot(kind='box', figsize=(10,8), color='r',yticks=range(0,70,5))
```

```python
sal = [15,22,24,25,30,23,34,43,19,20,52,21]

sal.sort()

sal
```

```python
 ser = pd.Series(sal)
 ser
```

```python
q1 = ser.quantile(0.25)
print('q1 =' , q1)

q3 = ser.quantile(0.75)
print('q3 =' , q3)

iqr = q3 - q1
print('iqr =' , iqr)
```

```python
# higher_side = q3 + (1.5 * iqr)

# lower_side = q1 - (1.5 * iqr)
```

```python
upside = q3 + (1.5 * iqr)

print(upside)
```

```python
lowside = q1 - (1.5 * iqr)

print(lowside)
```

```python
from pandas.plotting import scatter_matrix
```

```python
iris = sns.load_dataset('iris')

df = iris.drop(['species'],axis=1)

df.head()
```

```python
scatter_matrix(df, figsize=(10,8), color='r', diagonal='kde') #kde=kernel density  estimation

plt.show()
```

```python
# pie plot

rupees_monthly = [200,2530,900,15000,1500]
items = ['tuition_fees','mobile_bill','electicity_bill','rent','misc']

plt.pie(x=rupees_monthly, labels=items)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(x=rupees_monthly, labels=items, explode=[0,0,0,0.1,0], autopct='%.2f')
plt.show()
```

```python
# Numerical Data Plotting

# relplot
# scatterplot
# lineplot
```

```python
tips = sns.load_dataset('tips')

tips
```

```python
sns.relplot(x='total_bill',y='tip',data=tips)

plt.show()
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',style='time')

plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,size='size')

plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,size='size')

plt.show()
```

```python
sns.lineplot(x='total_bill',y='tip',data=tips)

plt.show()
```

```python
# Categorical Data Plotting

# catplot
# boxplot
# stripplot
# swarmplot
```

```python
sns.catplot(x='day',y='total_bill',data=tips)

plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)

plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='swarm',hue='size')
plt.show()
```

```python
sns.stripplot(x='size',y='total_bill',data=tips)

plt.show()
```

```python
sns.stripplot(x='smoker',y='total_bill',data=tips)

plt.show()
```
